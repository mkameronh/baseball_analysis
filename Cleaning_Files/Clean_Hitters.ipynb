```python
df_Batting.fillna(0, inplace=True)
```

```python
df_Batting_Grouped = df_Batting.groupby(['playerID']).sum().reset_index()
```

```python
df_Batting_Grouped = df_Batting_Grouped.rename(columns={ 'G':'Games', 
                                    'AB':'At_Bats', 
                                    'R':'Runs', 
                                    'H':'Hits', 
                                    '2B':'Doubles', 
                                    '3B':'Triples', 
                                    'HR':'Homeruns', 
                                    'RBI':'RBIs', 
                                    'SB':'Stolen_Bases', 
                                    'BB':'Walks', 
                                    'SO':'Strike_Outs', 
                                    'IBB':'Intentional_Walks', 
                                    'HBP':'Hit_by_Pitch', 
                                    'SH':'Sac_Hits', 
                                    'SF':'Sac_Flys', 
                                    'GIDP':'Double_Plays'
                                    })
```
