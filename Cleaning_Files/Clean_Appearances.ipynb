```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_People = pd.read_csv('..\Raw CSV\People.csv', encoding="ISO-8859-1")
df_AllStarFull = pd.read_csv('..\Raw CSV\AllstarFull.csv', encoding="ISO-8859-1")
df_Batting = pd.read_csv('..\Raw CSV\Batting.csv', encoding="ISO-8859-1")
df_Fielding = pd.read_csv('..\Raw CSV\Fielding.csv', encoding="ISO-8859-1")
df_Pitching = pd.read_csv('..\Raw CSV\Pitching.csv', encoding="ISO-8859-1")
df_HOF = pd.read_csv('..\Raw CSV\HallOfFame.csv', encoding= "ISO-8859-1")
df_Appearnces = pd.read_csv('..\Raw CSV\Appearances.csv', encoding= "ISO-8859-1")
```

```python
nulls = df_Appearnces.isnull().sum()
nulls[nulls > 0]
```

```python
df_Appearnces.fillna(0, inplace=True)
```

```python
df_Appearnces.drop(columns=['yearID',
                            'teamID',
                            'G_batting',
                            'G_defense',
                            'lgID',
                            'G_ph',
                            'G_pr'], 
                    axis=1, inplace=True)
```

```python
df_Appearnces_Grouped = df_Appearnces.groupby(['playerID']).sum().reset_index()
```

```python
df_Appearnces_Grouped
```

```python
df_Appearnces_Grouped.columns
```

```python
df_Appearnces_Grouped.rename(columns={'G_all':'Total_Games', 
                                      'GS':'Games_Started', 
                                      'G_p':'Pitcher', 
                                      'G_c':'Catcher', 
                                      'G_1b':'FirstBase', 
                                      'G_2b':'SecondBase', 
                                      'G_3b':'ThirdBase', 
                                      'G_ss':'ShortStop', 
                                      'G_lf':'LeftField',
                                      'G_cf':'CenterField', 
                                      'G_rf':'RightField', 
                                      'G_of':'Outfield', 
                                      'G_dh':'DH'
                                      }, inplace=True)
```

```python
df_Appearnces_Grouped
```

```python
df_Appearnces_Grouped.columns
```

```python
df_Appearnces_Grouped['Primary_Position'] = df_Appearnces_Grouped[['Pitcher', 
                                                                   'Catcher',
                                                                   'FirstBase', 
                                                                   'SecondBase', 
                                                                   'ThirdBase',
                                                                   'ShortStop', 
                                                                   'LeftField', 
                                                                   'CenterField',
                                                                   'RightField', 
                                                                   'Outfield', 
                                                                   'DH']].idxmax(axis=1)
```

```python
df_Appearnces_Grouped.columns
```

```python
df_Appearnces_Grouped.to_csv('..\Clean CSV\Clean_Positions.csv', index=False)
```
