```python
#Basic Hitting Statistics

#OBP - On base percentage -(H + BB + HBP) divided by (AB + BB + HBP + SF).
df_Batting1990['obp_numerator'] = (df_Batting1990['H'] + df_Batting1990['BB'] + df_Batting1990['HBP'])
df_Batting1990['obp_denominator'] = (df_Batting1990['AB'] + df_Batting1990['BB'] + df_Batting1990['HBP'] + df_Batting1990['SF'])
df_Batting1990['OBP'] = df_Batting1990['obp_numerator']/df_Batting1990['obp_denominator']
```

```python
# BA - batting average - hits divided by at bats
df_Batting1990['BA'] = df_Batting1990['H'] / df_Batting1990['AB']
```

```python
# TB - Total Bases. 1B, 2B, 3B, HR = 4
df_Batting1990['TB'] = df_Batting1990['H'] + 2*df_Batting1990['2B'] + 3*df_Batting1990['3B'] + 4 * df_Batting1990['HR']
```

```python
#SLG = TB / AB -> Slugging Average
df_Batting1990['SLG'] = df_Batting1990['TB'] / df_Batting1990['AB']
```

```python
# OPS = OBP + SLG
df_Batting1990['OPS'] = df_Batting1990['OBP'] + df_Batting1990['SLG']
```
