```python
df_People.drop(columns=['birthYear',
                        'birthCountry',
                        'birthMonth', 
                        'birthDay',
                        'birthState',
                        'birthCity',
                        'deathCountry',
                        'deathYear', 
                        'deathMonth', 
                        'deathDay', 
                        'deathState', 
                        'deathCity', 
                        'nameGiven', 
                        'retroID', 
                        'bbrefID',
                        'debut'], axis=1, inplace=True)
```

```python
# HOF Eligibility: Players who have played 10 seasons of Major League Baseball and retired for five full seasons.

#function to simplify dates with test
def clean_date(value):
    return value.split(' ')[0]
clean_date('1970-05-03 12:10:00')
```

```python
df_People['finalGame'] = df_People['finalGame'].astype('str').map(clean_date)
df_People.finalGame = pd.to_datetime(df_People['finalGame'], format='%Y-%m-%d')
df_People['Retired_Years']=(pd.datetime.today()-df_People['finalGame']).dt.days/365
```

```python
nulls = df_People.isnull().sum()
nulls[nulls > 0]
```

```python
weight_mean = df_People.weight.mean()
height_mean = df_People.height.mean()
```

```python
df_People['weight'].fillna(weight_mean, inplace=True)
df_People['height'].fillna(height_mean, inplace=True)
```

```python
df_People = pd.concat([df_People,pd.get_dummies(df_People.bats,drop_first=True)],axis=1)
```

```python
df_People = df_People.rename(columns={'L':'bats_L','R':'bats_R'})
```

```python
df_People
```

```python
pd.get_dummies(df_People.throws)
```

```python
df_People = pd.concat([df_People,pd.get_dummies(df_People.throws)],axis=1)
```

```python
df_People = df_People.rename(columns={'L':'throws_L','R':'throws_R'})
```

```python
df_People = df_People.drop(columns=['S','throws','bats'])
```

```python
df_People = df_People[['playerID',
                        'nameFirst', 
                        'nameLast', 
                        'weight', 
                        'height', 
                        'bats_L',
                        'bats_R',
                        'throws_L',
                        'throws_R', 
                        'finalGame', 
                        'Retired_Years', 
                        ]]
```
